```python
import numpy as np
df = df.astype({col: np.float64 for col in df.select_dtypes(include='int64').columns})

df.info()
```

1. Bar Chart **Top** 10 Countries by Number of Deliveries

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_countries = df['Country'].value_counts().nlargest(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries by Number of Deliveries')
plt.xlabel('Number of Deliveries')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

2. Line Plot  Deliveries Over Time

```python
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')
deliveries_over_time = df.groupby(df['Delivered to Client Date'].dt.to_period('M')).size()

deliveries_over_time.index = deliveries_over_time.index.to_timestamp()

plt.figure(figsize=(12, 6))
deliveries_over_time.plot()
plt.title('Deliveries Over Time (Monthly)')
plt.xlabel('Date')
plt.ylabel('Number of Deliveries')
plt.grid(True)
plt.tight_layout()
plt.show()
```

 3. Pie Chart Shipment Mode Proportions

```python
shipment_mode_counts = df['Shipment Mode'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(shipment_mode_counts, labels=shipment_mode_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Shipment Mode Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

4. Boxplot Line Item Value by Shipment Mode

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Shipment Mode', y='Line Item Value', palette='Set3')
plt.title('Line Item Value by Shipment Mode')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5.**Histogram** Monthly Delivery Count Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert date column to datetime if not already done
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')

# Extract year-month from the date
df['Month'] = df['Delivered to Client Date'].dt.to_period('M').astype(str)

# Plot histogram of deliveries per month
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Month', bins=len(df['Month'].unique()), color='steelblue', shrink=0.8)
plt.xticks(rotation=45)
plt.title('Monthly Delivery Count (Histogram)')
plt.xlabel('Month')
plt.ylabel('Number of Deliveries')
plt.tight_layout()
plt.show()
```

SUMMARY


The exploratory data analysis (EDA) of the SCSM Delivery History Dataset involved cleaning, transforming, and visualizing the data to extract meaningful insights from the delivery history of various pharmaceutical items. The dataset contained over 10,000 records with a mix of numeric, categorical, and date columns. Initial data cleaning steps included converting date fields to proper datetime formats, handling missing values, and transforming numeric columns that were incorrectly stored as objects.

I applied five different types of visualizations to understand the dataset from multiple perspectives:

Bar Chart – Top 10 Countries by Number of Deliveries:
This visualization helped identify which countries had the highest volume of deliveries. It showed that deliveries were concentrated in a few key countries, providing valuable information for regional supply chain analysis.

Histogram – Distribution of Line Item Values:
The histogram displayed the frequency distribution of line item values, giving insight into the most common cost ranges of deliveries. It revealed that most delivery values clustered in a lower range, with a long tail of higher-cost deliveries. This suggested the presence of a few high-value shipments and highlighted cost variability.

Pie Chart – Shipment Mode Proportions:
This chart provided a clear breakdown of the delivery methods used, such as air or sea freight. It helped in understanding logistics preferences and showed which shipment mode dominated the operations. This is crucial for assessing delivery speed, cost-efficiency, and planning logistics strategies.

Boxplot – Line Item Value by Shipment Mode:
The boxplot compared the distribution of delivery values across different shipment modes. It provided insights into cost differences and variability within each mode, helping to identify which methods were more expensive or had more inconsistent costs.

Histogram – Monthly Delivery Counts:
We visualized the number of deliveries per month using a time-based histogram. This revealed trends, seasonality, and any irregularities in delivery patterns over time. Such insights are important for demand forecasting and supply chain optimization.

In summary, the visualizations provided a comprehensive view of delivery behavior across time, geography, and shipment characteristics. The analysis highlighted cost patterns, shipment trends, and operational insights that can help stakeholders make informed decisions about logistics, inventory, and vendor relationships. Further analysis could include correlation heatmaps, vendor performance, and predictive modeling to optimize future deliveries.
